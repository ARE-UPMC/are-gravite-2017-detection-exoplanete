```python
%matplotlib notebook
from math import cos
from pylab import *
def vitesseradiale(v0,w,T,b):
    LR=[]
    i=0
    while i<20:
        LR.append(v0*w*cos((2*3.14*i/T)+b))
        i=i+0.5
    plot(range(0,4*T),LR, color="green",  linewidth=2, linestyle="-", label="vitesse radiale")
    legend(loc='lower left')
    show()
vitesseradiale(5.9,17.95,10,-1.6)
```

```python
%matplotlib notebook
import numpy as np
from math import cos
from matplotlib import *
def vitesseradiale(v0,w,T,b):
    LR=[]
    i=0
    while i<4:
        LR.append(v0*w*cos((2*np.pi*i/T)+b))
        i=i+0.5
    return LR

plot(range(0,8),vitesseradiale(5.9,17.95,10,-1.6), color="green",  linewidth=2, linestyle="-", label="vitesse radiale")
plot(range(0,8),vitesseradiale(5.9,18.08,10,-1.6), color="blue",  linewidth=2, linestyle="-", label="VNa1")
plot(range(0,8),vitesseradiale(5.9,17.81,10,-1.6), color="red",  linewidth=2, linestyle="-", label="VNa2")       

legend(loc='lower left')

show()
```

```python
%matplotlib notebook
import numpy as np
from pylab import *
def orbite(R):
    theta = linspace(0, 2*pi, 40)

    x = cos(theta)*R
    y = sin(theta)*R
    plot(x, y)

    show()
    
orbite(2)
```
